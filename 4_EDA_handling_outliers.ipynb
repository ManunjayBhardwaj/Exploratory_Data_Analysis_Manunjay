## 5 Number Summary and Box Plot

```python
# Taking out the 5 number summary
import numpy as np
list_marks=[45,32,76,34,77,46,54,44,68,33,64,78,32,44,75,35,90]
minimum,Q1,median,Q3,maximum=np.quantile(list_marks,[0,0.25,0.5,0.75,1])
print(minimum,Q1,median,Q3,maximum)
```

```python
# Calculating IQR
IQR=Q3 - Q1
print(IQR)
```

```python
# Calculating lower fence and higher fence
lower_fence=Q1-1.5*IQR
upper_fence=Q3+1.5*IQR
print(lower_fence,upper_fence)
```

```python
# Computing the box plot
import seaborn as sns
sns.boxplot(list_marks)
```

```python
# Adding an outlier
list_marks_1=[-100,45,32,76,34,77,46,54,44,68,33,64,78,32,44,75,35,90,200,230]
sns.boxplot(list_marks_1)
```
